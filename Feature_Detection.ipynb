```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

path = '/content/gdrive/My Drive/'
imgA = cv2.imread(path + 'Book_view0.png', cv2.IMREAD_GRAYSCALE)
imgB = cv2.imread(path + 'Book_view1.png', cv2.IMREAD_GRAYSCALE)

# create SIFT
sift = cv2.xfeatures2d.SIFT_create()

# key point detect
keypoints_A_sift = sift.detect(imgA, None)
keypoints_B_sift = sift.detect(imgB, None)

print('Keypoints of img A:', len(keypoints_A_sift))
print('Keypoints of img B:', len(keypoints_B_sift))

# draw key point
imgA_sift = cv2.drawKeypoints(imgA, keypoints_A_sift, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
imgB_sift = cv2.drawKeypoints(imgB, keypoints_B_sift, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# show
plt.figure(figsize=(20,20))
plt.subplot(211), plt.title('SIFT of img A'), plt.imshow(imgA_sift), plt.axis('off')
plt.subplot(212), plt.title('SIFT of img B'), plt.imshow(imgB_sift), plt.axis('off')
```

```python
# create SURF
surf = cv2.xfeatures2d.SURF_create()

# key point detect
keypoints_A_surf = surf.detect(imgA, None)
keypoints_B_surf = surf.detect(imgB, None)

print('Keypoints of img A:', len(keypoints_A_surf))
print('Keypoints of img B:', len(keypoints_B_surf))

# draw key point
imgA_surf = cv2.drawKeypoints(imgA, keypoints_A_surf, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
imgB_surf = cv2.drawKeypoints(imgB, keypoints_B_surf, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# show
plt.figure(figsize=(30,30))
plt.subplot(211), plt.title('SURF of img A'), plt.imshow(imgA_surf), plt.axis('off')
plt.subplot(212), plt.title('SURF of img B'), plt.imshow(imgB_surf), plt.axis('off')
```
